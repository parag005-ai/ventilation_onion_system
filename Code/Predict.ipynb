```python
import joblib
import pandas as pd

# Load model and columns
model = joblib.load("../Model/final_fan_speed_model.pkl")
columns = joblib.load("../Model/model_columns.pkl")

def predict_speed(gas, tmin, tmax, hmin, hmax):

    df = pd.DataFrame(
        [[gas, tmin, tmax, hmin, hmax]],
        columns=columns
    )

    return model.predict(df)[0]


# Example test
if __name__ == "__main__":

    test = predict_speed(920, 42, 48, 75, 82)

    print("Predicted Speed:", test)
```
